```python
from registration import *
from metadata import Metadata
import multiprocessing as mp
import os
import pickle
from hybescope_config.microscope_config import *
```

```python
from itertools import repeat
```

```python
md_pth = '/bigstore/Images2018/Robert/ca_celltypes/hybe_mcf10a_v2_noserum_2018Aug08_2018Aug18/'
md = Metadata(md_pth)

folder_name = 'hybes6to9'
save_dir = os.path.join(md_pth, folder_name)


if not os.path.exists(save_dir):
    os.makedirs(save_dir)
    
posnames = md.image_table.Position.unique()

hybe_list = ['hybe1', 'hybe2', 'hybe3', 'hybe4', 'hybe5']
hybe_list = ['hybe6', 'hybe7', 'hybe8', 'hybe9']
#hybe_list = ['hybe9']
fnames_dicts = [md.stkread(Channel='DeepBlue', Position=pos,
                           fnames_only=True, groupby='acq', 
                          hybe=hybe_list) for pos in posnames]

```

```python
def add_bead_data(fnames_dict, pos, ave_bead):
    bead_dict = {}
    for h in fnames_dict.keys():
        print(h)
        beads, _ = find_beads_3D(fnames_dict[h], ave_bead)
        bead_dict[h] = beads
    pickle.dump(bead_dict, open(os.path.join(save_dir, pos), 'wb'))
```

```python
os.environ['MKL_NUM_THREADS'] = '3'
os.environ['GOTO_NUM_THREADS'] = '3'
os.environ['OMP_NUM_THREADS'] = '3'


ncpu = 6
ppool = mp.Pool(ncpu)

```

```python
ppool.starmap(add_bead_data, zip(fnames_dicts, posnames, repeat(ave_bead, len(posnames))))
```
